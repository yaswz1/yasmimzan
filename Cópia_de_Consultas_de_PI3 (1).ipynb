```python
import sqlite3
from sqlite3 import Error

## Função de alteração (insert, update, delete...)
def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print(f"Query {query.split('(')[0].strip()} executada.")
        if cursor.rowcount != -1: print(f"{cursor.rowcount} linha(s) afetadas")

    except Error as e:
        print(f"Erro: '{e}'")


# Função de leitura de dados (select)
def execute_read_query(connection, query):
    cursor = connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()

        return result
    except Error as e:
        print(f"Erro: '{e}'")
```

* **Descrição da ação esperada ao rodar a consulta: Exibir todos os atendimentos com o nome do aluno, profissional responsável, data e descrição do prontuário.**

```python
query = """
DROP VIEW IF EXISTS vw_atendimentos;
"""
execute_query(connection, query)

query = """
CREATE VIEW vw_atendimentos AS
SELECT
    a.nome AS aluno,
    p.nome AS profissional,
    pr.dataProntuario,
    pr.descricao
FROM fazProntuario fp
JOIN Aluno a ON fp.RAAluno = a.RA
JOIN Profissional p ON fp.CPFProfissional = p.CPF
JOIN Prontuario pr ON fp.IDprontuario = pr.ID;
"""
execute_query(connection, query)
```

* **Justificativa para que a operação seja um Trigger (Contexto de PI3): O Trigger garante a integridade dos dados do sistema, impedindo o cadastro de prontuários com data futura. Essa validação ocorre automaticamente antes da inserção do registro.**
* **Descrição do evento que dispara o Trigger e das operações que ele executa: O Trigger é acionado antes da inserção de um novo prontuário. Caso a data informada seja maior que a data atual, a inserção é cancelada e uma mensagem de erro é exibida.**

```python
query = """
CREATE TRIGGER IF NOT EXISTS verificar_data
BEFORE INSERT ON Prontuario
FOR EACH ROW
WHEN NEW.dataProntuario > DATE('now')
BEGIN
    SELECT RAISE(ABORT, 'Data inválida');
END;
"""

execute_query(connection, query)
```
